```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
from matplotlib import rc
import pandas as pd
import os.path as path
from tqdm import tqdm
import scipy
from scipy.stats import linregress
from scipy.integrate import quad, dblquad
from scipy.optimize import curve_fit
from matplotlib import rcParams
rcParams['mathtext.fontset'] = 'stix'
rcParams['font.family'] = 'STIXGeneral'
```

```python
file1 = 'Volataje vs Conteos I=10.csv'
file2 = 'Intensidad vs Conteos V=650.csv'
Data1 = pd.read_csv(file1, sep=';')
Data2 = pd.read_csv(file2, sep=';')


Voltaje = np.float64(Data1['V'])
promC = np.float64(Data1['promC'])
sigmaC = np.float64(Data1['desvC'])
promA = np.float64(Data1['promA'])
sigmaA = np.float64(Data1['desvA'])


Intensidad = np.float64(Data2['I'])
sqrtProm = np.float64(Data2['SqrtProm'])
sigma = np.float64(Data2['Desv'])
```

```python
plt.xlabel(r'Voltaje [V]', fontsize=13, color='black')
plt.ylabel('Conteos por segundo', fontsize=13, color='black')
plt.tick_params(axis='both', colors='black')
plt.gca().spines['bottom'].set_color('black')
plt.gca().spines['top'].set_color('black')
plt.gca().spines['left'].set_color('black')
plt.gca().spines['right'].set_color('black')
plt.errorbar(Voltaje, promC, xerr = 0.5, yerr = sigmaC, fmt = '.', capsize=2, label = 'Obturador Cerrado', c='k')
plt.errorbar(Voltaje, promA, xerr = 0.5, yerr = sigmaA, fmt = '.', capsize=2, label = 'Obturador Abierto', c='r')
plt.legend()
```

```python
x = np.linspace(min(sigma), max(sigma), 100)
y = 1.1*x + 2
```

```python
plt.xlabel(r'Desviación estándar $(\sigma)$', fontsize=13, color='black')
plt.ylabel(r'Raíz del promedio $\left(\sqrt{\mu}\right)$', fontsize=13, color='black')
plt.tick_params(axis='both', colors='black')
plt.gca().spines['bottom'].set_color('black')
plt.gca().spines['top'].set_color('black')
plt.gca().spines['left'].set_color('black')
plt.gca().spines['right'].set_color('black')
plt.scatter(sigma, sqrtProm, c='k', label = 'Data')
plt.plot(x, y, label = r'$\sqrt{\mu}= (1.1 \pm 0.2)\; \sigma + (2\pm 1)$', color = 'r')
plt.legend()
```
